#Operadores Matemáticos


```python
# Soma
n1 = 10
n2 = 20
soma = n1 + n2
print(soma)

# Subtração
sub = n2 - n1
print(sub)

# Multiplicação
mult = n2 * n1
print(mult)

# Divisão
div = n2/n1
print(div)

# Incremento de variáveis
n1 += n1
print(n1)

# Potenciação
pot = 7 ** 7
pot2 = pow(7,7)
print(pot)
print(pot2)

pot = n1 ** n2
pot2 = pow(n1,n2)
print(pot)
print(pot2)

# Divisão resultando no valor inteiro
div2 = n1//n2
print(div2)


#Resto da divisão
resto = n1%n2
print(resto)
```

# Estruturas condicionais

```python
nota = 10
idade = 18

if nota >=7:
  print("Aprovado!")
else:
  print("Reprovado!")

if nota >=7 and nota <=9:
  print("Aprovado!")
elif nota > 9:
  print("Prodigio!")
else:
  print("Reprovado!")

pode_dirigir = "Pode dirigir" if idade >= 18 else "Não pode dirigir"
print(f"O resultado de sua analise deu: (pode_dirigir)")

match nota:
  case nota if nota >= 7 and nota <=9:
    print("Aprovado!")
  case nota if nota > 9:
    print("Prodigio!")
  case _:
    print("Reprovado!")

# Definição adicionada para 'opcao'
opcao = 0 # Exemplo: você pode definir 'opcao' com um valor diferente ou input do usuário

match opcao:
  case 0:
    print("Usuario não cadastrado")
  case 1:
    print("Usuario cadastrado com sucesso!")
  case 2:
    print("Usuario deletado com sucesso")
  case 3:
    print("Usuario editado com sucesso")
  case _:
    print("Opção invalida")


```
